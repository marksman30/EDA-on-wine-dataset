```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#load the data 
df=pd.read_csv("wine_dataset.csv")
```

```python
#shape of data 
df.shape
```

```python
#attributes in the dataset
df.columns
```

```python
#first few rows in the dataset
df.head()
```

```python
#summary of the dataframe
df.describe()
```

```python
#checking if there is missing data
df.isnull().any().any()#no null values
```

```python
#finding count of wines based on quality
sns.catplot(x='quality',kind='count',data=df)
```

```python
# plotting a correlogram 
sns.pairplot(df)
sns.plt.show()
```

```python
#correlogram in the form of heatmap
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),color="yellow",annot=True)
```

```python
plt.figure(figsize=(20,16))
for i,col in enumerate(list(df.columns.values)):
    plt.subplot(4,3,i+1)
    sns.distplot(df[col],color='g',kde=True,label='data')
    plt.grid()
    plt.legend(loc='upper right')
    plt.tight_layout()
    
```

```python
plt.figure(figsize=(10,15))
for i,col in enumerate(list(df.columns.values)):
    plt.subplot(4,3,i+1)
    df.boxplot(col)
    plt.grid()
    plt.tight_layout()
```
